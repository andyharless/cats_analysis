```python
vol_df = pd.DataFrame(pairs.data.short_name)
vol_df['volume'] = 999
for _, p in pairs.data.iterrows():
    token = p.short_name
    pair = Pair(pairs, token)
    vol = pair.get_volume()
    vol_df.loc[vol_df.short_name==token, 'volume'] = vol
vol_df.sort_values('volume', ascending=False)
```

```python
for _, p in pairs.data.iterrows():
    token = p.short_name
    pair = Pair(pairs, token)
    yld = pair.get_annualized_yield(lookback=30)
    thresh = pair.get_price_impact_threshold(impact=0.01)
    print(f'{token:6s}:  {yld: >7.2%},  {thresh: .2}')
```
